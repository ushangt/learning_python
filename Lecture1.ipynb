```python
# Iterator vs Comprehension 
L = [1, 2, 3, 4, 5]
for i in range(len(L)):
    L[i] += 10
print(L)
print("-----"*5)
print("-----"*5)
```

```python
def intersect(seq1, seq2): 
    res = []
    for x in seq1:
        if x in seq2:
            res.append(x) 
    return res
```

```python
s1 = "SPAM"
s2 = "SCAM"
y = intersect(s1, s2)
x = intersect([1, 2, 3], (1, 4))  
```
